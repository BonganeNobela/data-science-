### Section 2: Data Cleaning

##### 2.1 Check for any missing data (NaNs/nulls) and quantify the amount of missing data

````python
display(df.isna().sum().sort_values(ascending=False))
````

##### 2.2 Handling Missing Values

Let's handle the missing values using different methods.

##### 2.2.1 Filling missing numerical values with the mean

````python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
display(df.isna().sum().sort_values(ascending=False))
````

##### 2.2.2 Filling missing numerical values with the median

````python
df_median = pd.read_csv("/content/data/Airbnb_Data.csv")
numerical_cols = df_median.select_dtypes(include=['float64', 'int64']).columns
df_median[numerical_cols] = df_median[numerical_cols].fillna(df_median[numerical_cols].median())
display(df_median.isna().sum().sort_values(ascending=False))
````

##### 2.2.3 Filling missing categorical values with the mode

````python
df_mode = pd.read_csv("/content/data/Airbnb_Data.csv")

categorical_cols_with_na = df_mode.select_dtypes(include='object').columns[df_mode.select_dtypes(include='object').isnull().any()]

for col in categorical_cols_with_na:
    df_mode[col] = df_mode[col].fillna(df_mode[col].mode()[0])

display(df_mode.isna().sum().sort_values(ascending=False))
````

##### 2.2.4 Filling missing values with a constant value

````python
df_constant = pd.read_csv("/content/data/Airbnb_Data.csv")
df_constant['neighbourhood'] = df_constant['neighbourhood'].fillna('Unknown')
df_constant['thumbnail_url'] = df_constant['thumbnail_url'].fillna('No Image')
display(df_constant.isna().sum().sort_values(ascending=False))
````

##### 2.2.5 (**BONUS**) Impute missing values for the neighbourhood column using KNN

````python
df_neighbourhood = df.copy()

numerical_cols = df_neighbourhood.select_dtypes(include=['float64', 'int64']).columns.tolist() # we will use features like latitude, longitude etc to determine
numerical_cols.remove('id')

X_test_nbrs = df_neighbourhood.loc[df_neighbourhood["neighbourhood"].isna(), numerical_cols]
df_neighbourhood_not_na = df_neighbourhood.loc[
    ~df_neighbourhood["neighbourhood"].isna() # ~ is used for negation
]
X_train_nbrs, y_train_nbrs = (
    df_neighbourhood_not_na[numerical_cols],
    df_neighbourhood_not_na["neighbourhood"],
)

nbrs = KNeighborsClassifier(n_neighbors=2)
nbrs.fit(X_train_nbrs, y_train_nbrs)
nbrs_pred = nbrs.predict(X_test_nbrs)
nbrs_pred = pd.Series(nbrs_pred, index=X_test_nbrs.index)
nbrs_pred
````

````python
df["neighbourhood"] = df["neighbourhood"].fillna(value=nbrs_pred)
````

### Section 3: Categorical Encoding

##### 3.1 Identifying categorical columns

````python
categorical_cols = df.select_dtypes(include='object').columns
display(categorical_cols)
````

##### 3.2 Applying One-Hot Encoding

````python
cols_to_encode = ['property_type', 'room_type', 'bed_type']
df_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)
display(df_encoded.head())
display(df_encoded.shape)
````

````python
categorical_cols = ['property_type', 'room_type', 'bed_type']
encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
encoded_data = encoder.fit_transform(df[categorical_cols])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))
df_encoded = df.drop(columns=categorical_cols)
df_encoded = pd.concat([df_encoded, encoded_df], axis=1)
display(df_encoded.head())
````

### Section 4: Datetime Handling

##### 4.1 Convert `last_review` and `first_review` columns to datetime objects

````python
df['first_review'] = pd.to_datetime(df['first_review'], errors='coerce')
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
display(df[['first_review', 'last_review']].info())
````

##### 4.2 Capture the year and month from the `first_review` and `last_review` columns, use those columns to make other columns that include just the year and month from those two columns



````python
df['first_review_year'] = df['first_review'].dt.year
df['first_review_month'] = df['first_review'].dt.month
df['last_review_year'] = df['last_review'].dt.year
df['last_review_month'] = df['last_review'].dt.month
display(df[['first_review', 'last_review', 'first_review_year', 'first_review_month', 'last_review_year', 'last_review_month']].head())
````

##### 4.3 Filter for listings first reviewed after 2017

### Section 6: Feature Engineering

##### 6.1 Creating features from 'amenities'

Let's extract the number of amenities from the 'amenities' column.

````python
df['num_amenities'] = df['amenities'].apply(lambda x: len(x.replace('{','').replace('}','').split(',')) if x and x != '{}' else 0)
display(df[['amenities', 'num_amenities']].head())
````

##### 6.2 Create a feature for the length of the description. Name the column `description_length`

````python
df['description_length'] = df['description'].apply(lambda x: len(x) if isinstance(x, str) else 0)
display(df[['description', 'description_length']].head())
````

##### 6.3 Calculate the duration between the first and last review. Save them as `days_since_first_review`, `days_since_last_review` and `review_duration`

````python
df['first_review'] = pd.to_datetime(df['first_review'], errors='coerce')
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

df['days_since_first_review'] = (pd.to_datetime('today') - df['first_review']).dt.days
df['days_since_last_review'] = (pd.to_datetime('today') - df['last_review']).dt.days
df['review_duration'] = (df['last_review'] - df['first_review']).dt.days

display(df[['first_review', 'last_review', 'days_since_first_review', 'days_since_last_review', 'review_duration']].head())
````
