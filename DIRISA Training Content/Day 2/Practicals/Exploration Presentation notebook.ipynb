```python
median_bedrooms = df['total_bedrooms'].median()
df['total_bedrooms'].fillna(median_bedrooms, inplace=True) #filling missing values
```

```python
Q1 = df['median_house_value'].quantile(0.25)
Q2 = df['median_house_value'].quantile(0.50)
Q3 = df['median_house_value'].quantile(0.75)
IQR = Q3 - Q1

print('Q1 (25th percentile):', Q1)
print('Q2 (50th percentile):', Q2)
print('Q3 (75th percentile):', Q3)
print('IQR (Interquartile Range):', IQR)
```

```python
import numpy as np

# Calculate the lower and upper bounds using the IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Clip the 'median_house_value' column to the calculated bounds
df['median_house_value'] = np.clip(df['median_house_value'], lower_bound, upper_bound)

##Capping(Winsorizing)
##Keeps all rows but limits extreme values to the boundries
```

```python
print (df.shape)
```

```python
#Handle duplicates
print("Duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
```
